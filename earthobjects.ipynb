```python
from scipy import stats
z_scores = stats.zscore(df['estimated_diameter_max'])
threshold = 3
outliers = np.where(np.abs(z_scores) > threshold)
```

```python
df_clean = df.drop(outliers[0])
```

```python
df['diameter_to_distance_ratio'] = df['estimated_diameter_max'] / df['miss_distance']
df['velocity_to_distance_ratio'] = df['relative_velocity'] / df['miss_distance']
df['log_diameter'] = np.log(df['estimated_diameter_max'] + 1)  # Adiciona 1 para evitar logaritmo de zero
```

```python
bins = [0, 1, 2, np.inf]
labels = ['<1 LD', '1-2 LD', '>2 LD']
df['distance_bin'] = pd.cut(df['miss_distance'], bins=bins, labels=labels)
```
